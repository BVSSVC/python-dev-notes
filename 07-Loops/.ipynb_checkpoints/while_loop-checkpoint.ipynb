```python
a=str(1052)
smallest=int(a[0])
largest=int(a[0])
for i in a:
    digit_int=int(i)
    if digit_int>largest:
        largest=digit_int
    if digit_int<smallest:
        smallest=digit_int
print("Smallest:", smallest)
print("Largest:", largest)
```
