## Higher Order Functions

In Python functions are treated as first class citizens, allowing you to perform the following operations on functions:

- A function can take one or more functions as parameters
- A function can be returned as a result of another function
- A function can be modified
- A function can be assigned to a variable

In this section, we will cover:

1. Handling functions as parameters
2. Returning functions as return value from another functions
3. Using Python closures and decorators

### Function as a Parameter

````python
def sum_numbers(nums):
    return sum(nums)
def higher_order(f,lst):
    summation = f(lst)
    return summation
result = higher_order(sum_numbers,[1,2,3,4,5])
print(result)
````

### Function as a Return Value

````python
def square(x):
    return x**2
def cube(x):
    return x**3
def absolute(x):
    if x>0:
        return x
    else:
        return -(x)
def higher_order(type):
    if type=="square":
        return square
    elif type == 'cube':
        return cube
    elif type == 'absolute':
        return absolute
result = higher_order('square')
print(result(3))
result = higher_order('cube')
print(result(3))       # 27
result = higher_order('absolute')
print(result(-3))      # 3
````

# 🧭 **LEGB Scope in Python (Super Simple Notes)**

Python follows a rule called **LEGB** to decide **where to look for a variable**.

Think of it like searching for your lost keys — you check places in a specific order.

---

# 🔤 **L → Local**
- Variables inside the **current function**
- First place Python looks

**Example:**
```python
def func():
    x = 10  # Local
```

---

# 🔤 **E → Enclosing**
- Variables in **outer functions** (used in closures)
- Second place Python looks

**Example:**
```python
def outer():
    y = 20  # Enclosing
    def inner():
        print(y)
```

---

# 🔤 **G → Global**
- Variables defined at the **top level** of the file
- Third place Python looks

**Example:**
```python
z = 30  # Global
```

---

# 🔤 **B → Built‑in**
- Python’s built‑in names  
  (like `len`, `print`, `range`)

Last place Python checks.

---

# 🎯 **One-line summary**
**LEGB = Local → Enclosing → Global → Built‑in**  
Python searches in this order to find a variable.

---

# 🧸 **Child-level analogy**
Imagine you lost your pencil:

1. **Local** → check your desk  
2. **Enclosing** → check your classroom  
3. **Global** → check your house  
4. **Built‑in** → check the school office  

Python searches the same way.

## Python Closures

Python allows a nested function to access the outer scope of the enclosing function. This is is known as a Closure. Let us have a look at how closures work in Python. In Python, closure is created by nesting a function inside another encapsulating function and then returning the inner function. See the example below.

````python
def add_ten():
    ten = 10
    def add(nm):
        return nm+ten
    return add
closure_result = add_ten()
print(closure_result(5))
print(closure_result(20))
````

# 🧭 **Evaluation in Iterators — Simple Notes**

## 🟦 **1. What is Evaluation?**
**Evaluation** means *when* a value is actually produced or computed.

In iterators, Python uses:

### ⭐ **Lazy Evaluation**
Values are **not created immediately**.  
They are created **only when you ask for them**.

---

# 🟩 **2. Why Iterators Use Lazy Evaluation**
- Saves memory  
- Faster for large datasets  
- Works with infinite sequences  
- Avoids unnecessary computation  

---

# 🟧 **3. How Lazy Evaluation Works**
An iterator produces values **one at a time** using:

```python
next(iterator)
```

Each call to `next()` → evaluates the **next value**.

---

# 🟨 **4. Example Showing Evaluation**

```python
nums = iter([10, 20, 30])

print(next(nums))  # 10 (evaluated now)
print(next(nums))  # 20 (evaluated now)
print(next(nums))  # 30 (evaluated now)
```

Python does **not** evaluate all values at once.  
It evaluates **only when next() is called**.

---

# 🟪 **5. Iterator vs List (Evaluation Difference)**

| Type | Evaluation | When Values Are Created |
|------|------------|--------------------------|
| **List** | Eager | All at once |
| **Iterator** | Lazy | One at a time |

---

# 🧸 **6. Child-Level Analogy**
Think of a chai shop:

- You don’t get all cups at once  
- You get **one cup only when you ask**  
- The chaiwala makes it **fresh each time**

Iterator = chaiwala  
`next()` = asking for the next cup  
Lazy evaluation = chai made only when needed

---

# 🟫 **7. Why This Matters**
Lazy evaluation helps with:

- Large files  
- Streaming data  
- Infinite loops  
- Memory‑efficient programs

# 🔥 **Iterator vs Generator — The Real Difference**

## 🟦 1. **Iterator (manual machine)**

An **iterator** is an object that:

- Has `__iter__()`  
- Has `__next__()`  
- Gives values **one at a time**  
- Raises `StopIteration` when done  

To create an iterator, **you must write everything yourself**.

Example:

```python
class Counter:
    def __init__(self):
        self.num = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= 3:
            val = self.num
            self.num += 1
            return val
        raise StopIteration
```

This is long, manual, and you maintain the state yourself.

---

## 🟩 2. **Generator (automatic iterator)**

A **generator is just a simpler way to create an iterator**.

- Uses `yield`
- Python automatically creates:
  - `__iter__()`
  - `__next__()`
  - Internal state handling
  - StopIteration logic

Example:

```python
def my_gen():
    yield 1
    yield 2
    yield 3
```

This tiny function behaves exactly like the long iterator class above.

---

# 🟧 3. **Key Differences (Table)**

| Feature | Iterator | Generator |
|--------|----------|-----------|
| How created | Class with `__iter__` & `__next__` | Function with `yield` |
| Code length | Long | Very short |
| State handling | You manage manually | Python manages automatically |
| Memory usage | Can be high | Very low (lazy) |
| Ease of use | Harder | Much easier |
| Lazy evaluation | Yes | Yes |

---

# ⭐ One‑line summary

**Every generator *is* an iterator, but not every iterator is a generator.**  
Generators are simply **iterators made easy**.


## Python Decorators

A decorator is a design pattern in Python that allows a user to add new functionality to an existing object without modifying its structure. Decorators are usually called before the definition of a function you want to decorate.

### Creating Decorators

To create a decorator function, we need an outer function with an inner wrapper function.


````python
def my_decorator(func):
    def wrapper():
        print("####################")
        func()
        print("####################")
    return wrapper
def hello():
    print("Hello")
a=my_decorator(hello)
a()
````

````python
def upper_case(func):
    def wrapper():
        a = func()
        make_uppercase = a.upper()
        return make_uppercase
    return wrapper
@upper_case
def greeting():
    return "Welcome to python"
print(greeting())
````

````python
greeting()
````

````python
# First Decorator
def uppercase_decorator(function):
    def wrapper():
        func = function()
        make_uppercase = func.upper()
        return make_uppercase
    return wrapper

# Second decorator
def split_string_decorator(function):
    def wrapper():
        func = function()
        splitted_string = func.split()
        return splitted_string
    return wrapper

#Decorators will be executed from bottom to top
@split_string_decorator
@uppercase_decorator     # order with decorators is important in this case - .upper() function does not work with lists
def greeting():
    return 'Welcome to Python'
print(greeting())   # ['WELCOME', 'TO', 'PYTHON']
````

````python
def decorator_with_parameters(function):
    def wrapper_accepting_parameters(para1, para2, para3):
        function(para1, para2, para3)
        print("I live in {}".format(para3))
    return wrapper_accepting_parameters

@decorator_with_parameters
def print_full_name(first_name, last_name, country):
    print("I am {} {}. I love to teach.".format(
        first_name, last_name))

print_full_name("Asabeneh", "Yetayeh",'Finland')
````
